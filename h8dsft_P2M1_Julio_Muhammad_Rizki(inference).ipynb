```python
import pandas as pd
import numpy as np
import joblib
import tensorflow as tf
import warnings
warnings.filterwarnings(action='ignore')
```

```python
with open('preprocessor.pkl', 'rb') as file_1:
  preprocessor = joblib.load(file_1)
  
model_load = tf.keras.models.load_model('model.h5')
```

```python
membership_category_val = ['No Membership', 'Basic Membership', 'Premium Membership',
       'Gold Membership', 'Platinum Membership', 'Silver Membership']
joined_through_referral_val = ['Yes', 'No']
preferred_offer_types_val = ['Without Offers', 'Credit/Debit Card Offers',
       'Gift Vouchers/Coupons']
offer_application_preference_val = ['Yes', 'No']
feedback_val = ['Poor Website', 'Poor Customer Service', 'Too many ads',
       'Poor Product Quality', 'No reason specified',
       'Products always in Stock', 'Reasonable Price',
       'Quality Customer Care', 'User Friendly Website']

np.random.seed(20) 
membership_category = np.random.choice(membership_category_val, size=20)
joined_through_referral = np.random.choice(joined_through_referral_val, size=20)
preferred_offer_types = np.random.choice(preferred_offer_types_val, size=20)
avg_time_spent = np.random.randint(0,3235.578521, size=20)
avg_transaction_value = np.random.randint(800.460000,99914.050000, size=20)
avg_frequency_login_days = np.random.randint(0,73.061995,size=20)
points_in_wallet = np.random.randint(0,2069.069761,size=20)
offer_application_preference = np.random.choice(offer_application_preference_val, size=20)
feedback = np.random.choice(feedback_val, size=20)


data_inf = pd.DataFrame({'membership_category' : membership_category, 'joined_through_referral' : joined_through_referral,
       'preferred_offer_types' : preferred_offer_types, 'avg_time_spent' : avg_time_spent, 'avg_transaction_value' : avg_transaction_value,
       'avg_frequency_login_days': avg_frequency_login_days, 'points_in_wallet' : points_in_wallet,
       'offer_application_preference' : offer_application_preference, 'feedback' : feedback})
data_inf.head()
```

```python
data_inf_final = preprocessor.transform(data_inf)
data_inf['Churn Predict'] = np.round(model_load.predict(data_inf_final))
data_inf
```
